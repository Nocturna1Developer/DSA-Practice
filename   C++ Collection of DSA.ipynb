```python
int appearsOnce(int arr[]) {
    int n = sizeof(arr)/sizeof(arr[0])
    
    int res = arr[0] # must == arr[0] not 0
    for(int i = 0; i < n; i++)
    {
        res ^= arr[i]
    }
    return res
}
```

```python
int reverseArray(int arr[]) {
    int n = size(arr) / arr(arr[0]) 
    int start = 0
    int end = n - 1
    for(int i = 0; i < n; i++) {
        int temp = arr[start];
        arr[start] = arr[end];
        arr[end] = temp;
        start++;
        end--;
    }
}
```

```python
class Node {
    public:
        int data;
        Node* next;
}

class LinkedList {
    public:
        Node* head;
}

void insert(Node*, head_ref, int new_data) {
    Node* new_node = new Node();
    new_node->data = Node(new_data);
    (*head_ref) = new_node 
}


void insert_after(Node* prev_node, int new_data){
    if(prev_node == NULL) return;
    
    Node* new_node = new Node();
    new_node->data = new_data;
    new_node->next = prev_node->next;
    prev_node->next = new_node
        
}

int find(Node* head, int index)
{
    if(head == NULL) return -1;
    
    Node* new_node = new Node();
    int count = 0
    while(head != NULL){
        count++;
        if(count == index){
            return head->data;
        }
        head = head->next;
    }
}

int findFromEnd(Node* head, int index) {
    if(head == NULL) return -1
    
    Node* curr = new Node(); # forgot the pointer
    curr = head
    int length = 1
    
    while(curr != NULL){
        length++;
        curr = curr->next;
    }
    
    if(length < index) return;

    temp = head
    for(int i = 0; i < length - index + 1; i++){
        temp = temp->next;
    }
    return temp.data;
}


int findMiddle(Node* head){
    if(head == NULL) return -1;
    
    Node* curr = new Node();
    curr = head;
    length = 0;
    while(curr != NULL) {
        curr = curr->next;
        length++;
    }
    
    temp = head;
    for(int i = 0; i < length / 2; i++){
        temp = temp->next
    }
    return temp.data
    
}
```
