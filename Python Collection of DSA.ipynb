```python
# 8
def distinct(arr):
    s = set()
    for i in range(len(arr)):
        s.add(abs(arr[i]))
    return len(s)

arr = [-3, -2, 0, 3, 4, 5]
distinct(arr)
```

```python
# 11
def min_to_add(arr):
    res = 0
    min_num = 0
    for i in range(len(arr)):
        res += arr[i]
        if(res % 2 == 0):
            min_num = 2
        else:
            min_num = 1
    return min_num

arr1 = [1, 2, 3, 4, 5, 6, 7, 8]
print(min_to_add(arr1))
```

```python
import random
def makeRandomList(size):
    lyst = []
    for count in range(size):
        while True:
            # numbers in the list are unique and range from 1 through the size.
            number = random.randint(1, size)
            if not number in lyst: 
                lyst.append(number)
                break
    return sorted(lyst, reverse = True)


print(makeRandomList(89))
```
